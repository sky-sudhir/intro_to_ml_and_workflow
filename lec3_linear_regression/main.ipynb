```python
# change in one hot encoding for ocean_proximity with 0 and 1 values

ocean_proximity_dummies = pd.get_dummies(df['ocean_proximity'], prefix='ocean', drop_first=True).astype(int)
df = pd.concat([df, ocean_proximity_dummies], axis=1)



```

```python
df.head()

```

```python
# drop total_bedrooms ocean_proximity room_bed_ratio island columns
df=df.drop(columns=['total_bedrooms','ocean_proximity','room_bed_ratio','island'])
df.head()
```

```python
# min max scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
df.head()
```

```python
spearman_corr2 = df.corr(numeric_only=True,method='spearman')

# increase the size of the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(spearman_corr2,annot=True,cmap='coolwarm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error,root_mean_squared_error
X=df.drop(columns=['median_house_value'])
y=df['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
lr_r2score=r2_score(y_test,y_pred)
lr_mae=mean_absolute_error(y_test,y_pred)
lr_rmse=root_mean_squared_error(y_test,y_pred)
lr_mse=mean_squared_error(y_test,y_pred)
```

```python
# Compare with a Ridge regression model (alpha=1.0)
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_ridge_pred = ridge_model.predict(X_test)
```

```python
ridge_model.coef_
```

```python
ridge_model.intercept_
```

```python
ridge_mse=mean_squared_error(y_test,y_ridge_pred)
ridge_r2_score=r2_score(y_test,y_ridge_pred)
ridge_mae=mean_absolute_error(y_test,y_ridge_pred)
ridge_rmse=root_mean_squared_error(y_test,y_ridge_pred)
```

```python
print("================= Linear Regression Metrics vs Ridge Regression Metrics (alpha=1.0) ================ ")
print(lr_r2score, ridge_r2_score)
print(lr_mae, ridge_mae)
print(lr_rmse, ridge_rmse)
print(lr_mse, ridge_mse)
```
